## Descrição do Código usando SVD

O código implementa um **sistema de recomendação de filmes** usando **SVD (Singular Value Decomposition)**.

1. **Filtragem inicial**  
   - Mantém apenas usuários e filmes com **50+ avaliações** para melhorar a qualidade dos dados.  
   - Cria uma matriz **usuário × filme** onde cada célula representa a nota atribuída.  

2. **Divisão dos dados**  
   - Separa em **treino (80%)** e **teste (20%)** para permitir avaliação do modelo.  

3. **Treinamento com SVD**  
   - Preenche notas faltantes com **0**.  
   - Aplica **SVD truncado** com 20 fatores latentes para identificar padrões de preferência.  

4. **Previsão de notas**  
   - Reconstrói a matriz de avaliações previstas multiplicando os fatores de usuários e filmes.  

5. **Função de recomendação**  
   - Para um usuário, seleciona filmes não avaliados, ordena pelas notas previstas e retorna os **Top 10** recomendados.


```python
# Criar uma tabela de gêneros usando one-hot encoding
genres_dummies = movies_df['genres'].str.get_dummies(sep='|')

# Adicionar a tabela de gêneros ao DataFrame de filmes
movies_df = pd.concat([movies_df, genres_dummies], axis=1)

# Remover a coluna original 'genres'
movies_df = movies_df.drop('genres', axis=1)
```

```python
display(movies_df.head(10))
```

```python
# Junta as tabelas de avaliações (ratings_df) e de filmes (movies_df) usando 'movieId' como chave
df_merged = pd.merge(ratings_df, movies_df, on='movieId')
# Conta quantas avaliações cada usuário fez
user_counts = df_merged['userId'].value_counts()
print("Número de avaliações por usuário:")
display(user_counts)
# Conta quantas avaliações cada filme recebeu
movie_counts = df_merged['movieId'].value_counts()
print("\nNúmero de avaliações por filme:")
display(movie_counts)
```

```python
# Mantém apenas os usuários que avaliaram 50 ou mais filmes
df_merged = df_merged[df_merged['userId'].isin(user_counts[user_counts >= 50].index)]
# Mantém apenas os filmes que receberam 50 ou mais avaliações
df_merged = df_merged[df_merged['movieId'].isin(movie_counts[movie_counts >= 50].index)]
# Exibe as primeiras linhas do DataFrame filtrado
display(df_merged.head())
```

```python
# Cria uma matriz de usuários x filmes:
# - 'index' define que cada linha será um usuário (userId)
# - 'columns' define que cada coluna será um filme (movieId)
# - 'values' define que o valor na célula será a nota dada (rating)
user_item_matrix = df_merged.pivot(index='userId', columns='movieId', values='rating')
display(user_item_matrix.head(10))
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df_merged, test_size=0.2, random_state=42)

print("Shape of train_data:", train_data.shape)
print("Shape of test_data:", test_data.shape)
```

```python
# Cria a matriz usuário-itens para os dados de treino:
# - 'index' define que cada linha será um usuário (userId)
# - 'columns' define que cada coluna será um filme (movieId)
# - 'values' define que os valores da matriz serão as notas (rating)
# - 'fillna(0)' substitui valores ausentes por 0 (usuário não avaliou o filme)
train_user_item_matrix = train_data.pivot_table(index='userId', columns='movieId', values='rating').fillna(0)

# Cria o modelo SVD truncado:
# - 'n_components=20' define o número de fatores latentes
# - 'random_state=42' garante reprodutibilidade dos resultados
svd = TruncatedSVD(n_components=20, random_state=42)

# Aplica o SVD na matriz usuário-itens para decompor em fatores latentes
latent_matrix = svd.fit_transform(train_user_item_matrix)

# Exibe as dimensões da matriz de fatores latentes (usuários x fatores)
print("Shape of the latent factor matrix:", latent_matrix.shape)
```

```python
# Reconstrói a matriz de avaliações previstas multiplicando:
# - a matriz de fatores latentes dos usuários (latent_matrix)
# - pelos componentes do SVD (svd.components_) que representam os itens
reconstructed_ratings = np.dot(latent_matrix, svd.components_)

# Converte a matriz reconstruída para um DataFrame:
# - index = usuários
# - columns = filmes
reconstructed_ratings_df = pd.DataFrame(
    reconstructed_ratings,
    index=train_user_item_matrix.index,
    columns=train_user_item_matrix.columns
)

def recommend_movies(user_id, reconstructed_ratings_df, n_recommendations=10):
    """
    Recomenda filmes para um determinado usuário com base nas classificações previstas.

    Args:
        user_id (int): O ID do usuário.
        reconstructed_ratings_df (pd.DataFrame): DataFrame com as notas previstas.
        n_recommendations (int): Quantidade de recomendações a retornar.

    Returns:
        pd.DataFrame: DataFrame com os N filmes recomendados.
    """
    # Obtém as notas originais do usuário (treinamento)
    user_original_ratings = train_user_item_matrix.loc[user_id]

    # Obtém as notas previstas para o usuário
    user_predicted_ratings = reconstructed_ratings_df.loc[user_id]

    # Filtra apenas filmes que o usuário ainda não avaliou (nota original = 0)
    unrated_movies = user_original_ratings[user_original_ratings == 0].index

    # Ordena os filmes não avaliados pela nota prevista (do maior para o menor)
    recommendations = user_predicted_ratings[unrated_movies].sort_values(ascending=False)

    # Seleciona os N filmes do topo
    top_recommendations = recommendations.head(n_recommendations)

    # Junta as informações dos filmes (id + título) a partir do movies_df
    recommended_movies_info = movies_df[movies_df['movieId'].isin(top_recommendations.index)][['movieId', 'title']]

    return recommended_movies_info
```

```python
# Exemplo: pega um usuário aleatório
random_user = random.choice(train_user_item_matrix.index)

# Gera as recomendações para esse usuário
recommendations = recommend_movies(random_user, reconstructed_ratings_df)

# Mostra na tela as recomendações
print(f"Top 10 movie recommendations for user {random_user}:")
display(recommendations)
```

```python
# Obtém as notas previstas para os dados de teste:
# Para cada linha do test_data, pega a nota prevista correspondente da matriz reconstruída
test_predictions = test_data.apply(
    lambda row: reconstructed_ratings_df.loc[row['userId'], row['movieId']],
    axis=1
)

# Calcula o erro quadrático para cada previsão:
# (nota real - nota prevista)²
squared_errors = (test_data['rating'] - test_predictions) ** 2

# Calcula a média dos erros quadráticos
mean_squared_error = squared_errors.mean()

# Tira a raiz quadrada da média para obter o RMSE
rmse = np.sqrt(mean_squared_error)

# Mostra o valor final do RMSE formatado com 4 casas decimais
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# 1. Seleciona um usuário aleatório do conjunto de treinamento
random_user = random.choice(train_user_item_matrix.index)
print(f"Selected random user: {random_user}")

# 2. Filtra os filmes bem avaliados por esse usuário (nota >= 4.0)
highly_rated_movies = df_merged[
    (df_merged['userId'] == random_user) &  # apenas o usuário selecionado
    (df_merged['rating'] >= 4.0)            # notas maiores ou iguais a 4.0
]

# Exibe os filmes com as respectivas notas
print("\nMovies rated 4.0 or higher by this user:")
display(highly_rated_movies[['title', 'rating']])

# 3. Gera as 10 principais recomendações de filmes para o usuário
recommendations = recommend_movies(random_user, reconstructed_ratings_df, 10)

# Exibe a lista de recomendações
print("\nTop 10 movie recommendations:")
display(recommendations)
```

## Descrição do Código usando REGRESSÂO LOGISTICA

O código implementa um **sistema de recomendação** utilizando **Regressão Logística** para prever se um usuário gostará de um filme.

1. **Transformação da variável-alvo**  
   - Cria a coluna binária `liked`: valor **1** se a nota ≥ 4 e **0** caso contrário.

2. **Preparação dos dados**  
   - Define como *features* o `userId`, `movieId` e as colunas de gênero do filme.  
   - Converte `userId` e `movieId` para string e aplica **One-Hot Encoding**.  
   - Divide os dados em treino (80%) e teste (20%), mantendo a proporção da variável-alvo.

3. **Treinamento do modelo**  
   - Utiliza **Regressão Logística** (`solver='liblinear'`) para classificação binária.  
   - Treina o modelo nos dados de treino.

4. **Avaliação do modelo**  
   - Mede **Acurácia** e **AUC-ROC** no conjunto de teste para verificar desempenho.

5. **Função de recomendação**  
   - Para um usuário, identifica filmes não avaliados.  
   - Calcula a probabilidade de “gostar” usando o modelo treinado.  
   - Ordena por probabilidade e retorna os **Top 10** com título e pontuação prevista.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
# 1. Transformar as notas em classes (binária: 1 se nota >= 4, 0 se nota < 4)
# Cria uma nova coluna 'liked' que é 1 se a nota for >= 4, e 0 caso contrário.
df_merged['liked'] = (df_merged['rating'] >= 4.0).astype(int)

# 2. Preparar os dados para o modelo de Regressão Logística
# As features serão o userId, o movieId, e as colunas de gênero.
genre_columns = movies_df.columns.drop(['movieId', 'title', 'year']) # Obtém as colunas de gênero do movies_df
features = ['userId', 'movieId'] + list(genre_columns) # Combina as features de usuário, filme e gênero
target = 'liked'

X = df_merged[features].copy() # Cria uma cópia explícita para evitar o aviso SettingWithCopyWarning
y = df_merged[target]

# Converte userId e movieId para string para aplicar one-hot encoding
X['userId'] = X['userId'].astype(str)
X['movieId'] = X['movieId'].astype(str)

# Aplica one-hot encoding em userId e movieId, mantendo as outras features
X_encoded = pd.get_dummies(X, columns=['userId', 'movieId'], drop_first=True)

# Divide os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

# Usa os dados originais X_train e X_test diretamente
X_train_model = X_train
X_test_model = X_test

# 3. Treinar o modelo de Regressão Logística
# Inicializa o modelo de Regressão Logística
model = LogisticRegression(random_state=42, solver='liblinear') # 'liblinear' é adequado para datasets menores e classificação binária

# Treina o modelo com os dados de treino
model.fit(X_train_model, y_train)

# 4. Avaliar o modelo nos dados de teste
# Faz previsões nos dados de teste
y_pred = model.predict(X_test_model)
y_prob = model.predict_proba(X_test_model)[:, 1] # Probabilidade da classe positiva (liked=1)

# Calcula a acurácia e o AUC-ROC
accuracy = accuracy_score(y_test, y_pred)
auc_roc = roc_auc_score(y_test, y_prob)

print(f"Accuracy on test set: {accuracy:.4f}")
print(f"AUC-ROC on test set: {auc_roc:.4f}")
```

```python
# Função para recomendar filmes para um usuário
def recommend_movies_lr(user_id, model, movies_df, features, n_recommendations=10):
    """
    Recomenda filmes para um determinado usuário usando o modelo de Regressão Logística.

    Args:
        user_id (int): O ID do usuário.
        model: O modelo de Regressão Logística treinado.
        movies_df (pd.DataFrame): DataFrame com informações dos filmes.
        features (list): Lista de features usadas no treinamento do modelo.
        n_recommendations (int): Quantidade de recomendações a retornar.

    Returns:
        pd.DataFrame: DataFrame com os top N filmes recomendados e suas probabilidades previstas de 'gostar'.
    """
    # Obtém todos os IDs de filmes avaliados pelo usuário no dataframe original
    rated_movie_ids = df_merged[df_merged['userId'] == user_id]['movieId'].unique()

    # Obtém todos os filmes não avaliados pelo usuário
    unrated_movies_df = movies_df[~movies_df['movieId'].isin(rated_movie_ids)].copy()

    if unrated_movies_df.empty:
        print(f"Usuário {user_id} avaliou todos os filmes disponíveis.")
        return pd.DataFrame()

    # Prepara os dados para predição
    # Cria um DataFrame com user_id e movieId para filmes não avaliados
    predict_data = unrated_movies_df[['movieId'] + list(genre_columns)].copy()
    predict_data['userId'] = user_id

    # Seleciona as features e converte userId e movieId para string
    X_predict = predict_data[features].copy()
    X_predict['userId'] = X_predict['userId'].astype(str)
    X_predict['movieId'] = X_predict['movieId'].astype(str)

    # Aplica one-hot encoding em userId e movieId nos dados de predição
    X_predict_encoded = pd.get_dummies(X_predict, columns=['userId', 'movieId'])

    # Reindexa os dados de predição para corresponder às colunas dos dados de treino, preenchendo ausentes com 0
    X_predict_encoded = X_predict_encoded.reindex(columns=X_train_model.columns, fill_value=0)

    # Prediz a probabilidade de "gostar" para cada filme não avaliado
    unrated_movies_df['predicted_liked_prob'] = model.predict_proba(X_predict_encoded)[:, 1]

    # Ordena os filmes pela probabilidade prevista (decrescente)
    recommendations = unrated_movies_df.sort_values(by='predicted_liked_prob', ascending=False)

    # Seleciona os top N filmes e inclui o título e a probabilidade prevista
    top_recommendations = recommendations.head(n_recommendations)[['movieId', 'title', 'predicted_liked_prob']]

    return top_recommendations
```

```python
# Exemplo: Recomendar filmes para um usuário aleatório
random_user = random.choice(df_merged['userId'].unique())
print(f"\nRecomendando filmes para o usuário {random_user}:")

# Passa a lista de features para a função de recomendação
recommendations_lr = recommend_movies_lr(random_user, model, movies_df, features)

if not recommendations_lr.empty:
    print("\nTop 10 recomendações de filmes (Regressão Logística):")
    display(recommendations_lr)
```

## Descrição do Código usando REDES NEURAL

código implementa um recomendador usando **rede neural** no TensorFlow/Keras para prever se um usuário gostará de um filme.

1. **Transformação da variável-alvo**  
   - Coluna `liked`: 1 se nota ≥ 4, caso contrário 0.

2. **Preparação dos dados**  
   - Mapeamento de `userId` e `movieId` para índices numéricos (necessário para embeddings).  
   - Entrada do modelo:  
     - `user_idx` (ID do usuário)  
     - `movie_idx` (ID do filme)  
     - Vetor de gêneros (`genres`)  

3. **Arquitetura da Rede Neural**  
   - **Camada de Embedding** para usuários e filmes (32 dimensões cada).  
   - **Concatenação** dos embeddings + gêneros.  
   - Duas **camadas densas**:  
     - 64 neurônios (ReLU)  
     - 32 neurônios (ReLU)  
   - Saída: 1 neurônio (sigmoid) → probabilidade de gostar.

4. **Treinamento**  
   - Otimizador: `Adam`  
   - Função de perda: `Binary Crossentropy`  
   - Métricas: `Accuracy` e `AUC`  
   - Épocas: 5 | Batch size: 256

5. **Recomendação**  
   - Seleciona filmes não avaliados pelo usuário.  
   - Calcula a probabilidade de “gostar” usando o modelo.  
   - Ordena e retorna os **Top N** filmes com título e probabilidade.

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from sklearn.model_selection import train_test_split
```

```python
# 1. Criar a coluna alvo binária
df_merged['liked'] = (df_merged['rating'] >= 4.0).astype(int)

# 2. Preparar dados
genre_columns = movies_df.columns.drop(['movieId', 'title', 'year'])
target = 'liked'

# Dividir treino e teste
train_df, test_df = train_test_split(df_merged, test_size=0.2, random_state=42, stratify=df_merged[target])

# Mapeamento para IDs consecutivos (necessário para embeddings)
user_map = {u: i for i, u in enumerate(df_merged['userId'].unique())}
movie_map = {m: i for i, m in enumerate(df_merged['movieId'].unique())}

train_df['user_idx'] = train_df['userId'].map(user_map)
train_df['movie_idx'] = train_df['movieId'].map(movie_map)
test_df['user_idx'] = test_df['userId'].map(user_map)
test_df['movie_idx'] = test_df['movieId'].map(movie_map)

# Entradas
X_train_user = np.array(train_df['user_idx'])
X_train_movie = np.array(train_df['movie_idx'])
X_train_genres = np.array(train_df[genre_columns])
y_train = np.array(train_df[target])

X_test_user = np.array(test_df['user_idx'])
X_test_movie = np.array(test_df['movie_idx'])
X_test_genres = np.array(test_df[genre_columns])
y_test = np.array(test_df[target])

# 3. Criar a rede neural com embeddings
num_users = len(user_map)
num_movies = len(movie_map)
num_genres = len(genre_columns)

# Inputs
user_input = layers.Input(shape=(1,), name="userId")
movie_input = layers.Input(shape=(1,), name="movieId")
genres_input = layers.Input(shape=(num_genres,), name="genres")

# Embeddings
user_emb = layers.Embedding(num_users, 32)(user_input)
movie_emb = layers.Embedding(num_movies, 32)(movie_input)

# Flatten
user_vec = layers.Flatten()(user_emb)
movie_vec = layers.Flatten()(movie_emb)

# Concatenar tudo
concat = layers.Concatenate()([user_vec, movie_vec, genres_input])

# Camadas densas
dense1 = layers.Dense(64, activation='relu')(concat)
dense2 = layers.Dense(32, activation='relu')(dense1)
output = layers.Dense(1, activation='sigmoid')(dense2)

# Modelo
model = Model(inputs=[user_input, movie_input, genres_input], outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC()])
model.summary()

# 4. Treinar
model.fit([X_train_user, X_train_movie, X_train_genres], y_train,
          validation_data=([X_test_user, X_test_movie, X_test_genres], y_test),
          epochs=5, batch_size=256)
```

```python
# 5. Função de recomendação com rede neural
def recommend_movies_nn(user_id, model, movies_df, genre_columns, n_recommendations=10):
    # Verifica se o usuário existe no mapeamento
    if user_id not in user_map:
        print(f"Usuário {user_id} não está no conjunto de treino. Não é possível recomendar.")
        return pd.DataFrame()

    # Filmes que o usuário já avaliou
    rated_movie_ids = df_merged[df_merged['userId'] == user_id]['movieId'].unique()

    # Obtém todos os filmes não avaliados pelo usuário
    unrated_movies_df = movies_df[~movies_df['movieId'].isin(rated_movie_ids)].copy()

    # Filtra apenas filmes que estão no mapeamento do treinamento
    unrated_movies_df = unrated_movies_df[unrated_movies_df['movieId'].isin(movie_map.keys())]


    if unrated_movies_df.empty:
        print(f"Usuário {user_id} avaliou todos os filmes conhecidos pelo modelo.")
        return pd.DataFrame()

    # Cria os vetores de entrada para a rede neural
    user_idx = np.full(len(unrated_movies_df), user_map[user_id], dtype=np.int32)
    movie_idx = unrated_movies_df['movieId'].map(movie_map).astype(np.int32).values
    genres_data = unrated_movies_df[genre_columns].values.astype(np.float32)

    # Previsões
    probs = model.predict([user_idx, movie_idx, genres_data], verbose=0).flatten()

    # Adiciona a coluna de probabilidade
    unrated_movies_df['predicted_liked_prob'] = probs

    # Ordena pelas maiores probabilidades
    recommendations = unrated_movies_df.sort_values(by='predicted_liked_prob', ascending=False)

    return recommendations.head(n_recommendations)[['movieId', 'title', 'predicted_liked_prob']]
```

```python

# 6. Exemplo de uso
random_user = random.choice(df_merged['userId'].unique())
print(f"\nRecomendações para o usuário {random_user}:")
recommendations_nn = recommend_movies_nn(random_user, model, movies_df, genre_columns)
print(recommendations_nn)
```
